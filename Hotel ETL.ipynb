```python
# Load the data to the sql server 
import sqlalchemy as sal

# Define your PostgreSQL details
username = 'postgres'
password = 'Edna90!'
hostname = 'localhost'
port = '5432'
```

```python
# Create the SQLAlchemy engine
engine = sal.create_engine(f'postgresql+psycopg2://{username}:{password}@{hostname}:{port}/postgres')

# Write hotel data for each year to hotel_table
hotel_data_2018.to_sql('table_2018', con=engine, if_exists='append', index=False)

```

```python
hotel_data_2019.to_sql('table_2019', con=engine, if_exists='append', index=False)
```

```python
hotel_data_2020.to_sql('table_2020', con=engine, if_exists='append', index=False)
```

```python
# Write meal_cost and market_segment data
meal_cost.to_sql('meal_costs', con=engine, if_exists='replace', index=False)
```

```python
market_segment.to_sql('market_segments', con=engine, if_exists='replace', index=False)
```
